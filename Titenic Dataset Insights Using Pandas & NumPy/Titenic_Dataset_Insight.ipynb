```python
# Importing the necessary library
import pandas as pd
import numpy as np
```

```python
# 0. Data load

# Dataset loading
df = pd.read_csv('D:\Project\Titenic Dataset Insights Using Pandas & NumPy\Titanic-Dataset.csv') # Convert the DataFrame into the CSV
print(df.head()) # First fice rows shows
```

Explanation:

pd.read_csv() → loads CSV into a DataFrame

head() → shows first 5 rows

```python
# 1. Basic Data Exploration (Pandas)
# (A) Shape, column, info

print('Missing values in each column')
df.shape # row, columns
df.columns # column names
df.info() # datatypes + null count
```

Explanation:

shape → (rows, columns)

columns → list of column names

info() → data types + non-null counts

```python
# (B) Summary Statistics

df.describe() # States of numeric columns: mean, std, min/max
df.describe(include="object") # Stats of column object: unique, top, freq
```

Explanation:

describe() → numeric stats (mean, std, min, max)

include="object" → categorical stats (unique, top value)

```python
# 2. Missing Values (Null Handinling)
# (A) Count Missing Values
df.isna().sum().sort_values(ascending=False)
```

Explanation:

isna() → checks null values

sum() → counts missing

sort_values() → sorts highest missing first

```python
# (B) Missing Percentage
(df.isna().mean() * 100).sort_values(ascending=False)
```

```python
# Selecting Data
# (A) Select Columns

df["Age"].head()
df[["Age","Sex", "Survived"]].head()
```

```python
# (B) Filter Rows(Conditions)
df[df["Sex"] == "female"].head()
df[(df["Age"] > 30) & (df["Pclass"] == 1)].head()
```

Explanation:

& → AND condition

| → OR condition

```python
# 4. Sorting and Value Counts

df.sort_values("Fare", ascending=False).head(10)

df["Sex"].value_counts()

df["Pclass"].value_counts(normalize=True) * 100
```

Explanation:

sort_values() → sorts rows

value_counts() → category frequency

normalize=True → percentage

```python
# 5. GroupBy
# (A) Survival Rate by Gender

df.groupby("Sex")["Survived"].mean()
```

Explanation:

Since Survived = 0 or 1

Mean = survival rate

```python
# (B) Survival by Class and Gender

df.groupby(["Pclass", "Sex"])["Survived"].mean()
```

```python
# (C) Multiple Aggregations

df.groupby("Pclass").agg(
    avg_fare=("Fare", "mean"),
    max_fare=("Fare", "max"),
    total_passengers=("PassengerId", "count")
)
```

```python
# 6️. Creating New Features
# (A) Create Age Groups

df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 35, 60, 100],
    labels=["Child", "Teen", "Young", "Adult", "Senior"]
)

df[["Age", "AgeGroup"]].head()
```

Explanation:

pd.cut() → converts continuous values into categories

```python
# (B) Create Family Size

df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df[["SibSp", "Parch", "FamilySize"]].head()
```

```python
# 7️. Handling Missing Values
# (A) Fill Age with Median

df["Age"] = df["Age"].fillna(df["Age"].median())
```

Why median?

Less affected by outliers

```python
# (B) Fill Embarked with Mode

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
# 8️. NumPy Practice on Titanic
# (A) Convert to NumPy Array

ages = df["Age"].to_numpy()
fares = df["Fare"].to_numpy()

ages[:10]
```

```python
# (B) Mean, Median, Std

np.mean(ages)
np.median(ages)
np.std(ages)

print(np.mean(ages))
print(np.median(ages))
print(np.std(ages))
```

```python
# (C) Find Passengers with Fare Above Average

avg_fare = np.mean(fares)

indices = np.where(fares > avg_fare)[0]

df.iloc[indices][["Name", "Fare"]].head()
```

Explanation:

np.where() → returns index positions

iloc → index-based selection

```python
# (D) Min-Max Normalization

fare_min = fares.min()
fare_max = fares.max()

fare_norm = (fares - fare_min) / (fare_max - fare_min)

fare_norm[:10]
```

```python
# 9️. Pivot Table

pd.pivot_table(
    df,
    values="Survived",
    index="Sex",
    columns="Pclass",
    aggfunc="mean"
)
```

Explanation:

Shows survival rate matrix by Gender vs Class
